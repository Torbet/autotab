```python
import sys

sys.path.append('..')
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast

%matplotlib inline
```

```python
# data = np.load('../data/test_data_comp.npz')
data = np.load('../data/raw/audio_tabs_batch_1.npz')
tabs = data['tabs']
audios = data['audio']
```

```python
for i in range(5):
  audio = audios[i]

  print(audio.shape)  # 80, 3000

  plt.figure(figsize=(10, 5))
  sns.heatmap(audio, cmap='viridis')
  plt.title('Audio')
  plt.show()
```

```python
# confusion matrix on test set
with open('../results/tiny_confusion.txt', 'r') as f:
  confusion = ast.literal_eval(f.read())

confusion = np.array(confusion)
confusion = confusion / np.sum(confusion, axis=1)[:, np.newaxis]

plt.figure(figsize=(10, 10))
sns.heatmap(confusion, annot=False, cmap='viridis')
```
